## import libraries

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('INT-lasAnimasCO.xlsx')
```

```python
data
```

### Removing Duplicated Values in ('Mail Address','Mail City','Mail State','Mail Zip') Columns

```python
data = data.drop_duplicates(['Mail Address','Mail City','Mail State','Mail ZIP'])
```

### Removing Null Values in ('Mail Address', 'Mail City', 'Mail State', 'Mail Zip')

```python
data.isnull().sum()
```

```python
data.dropna(subset=['Mail Address','Mail City','Mail State','Mail ZIP','Primary Name'], inplace=True)
```

```python
# Function to check if a string ends with any of the specified substrings
def ends_with_specific_strings(s):
    words = ['company', 'inc', 'llc', 'state','land']
    return any(s.lower().endswith(word) for word in words)

# Apply the filter
mask = data['Primary Name'].apply(lambda x: not ends_with_specific_strings(x))
data = data[mask]

data
```

```python
data.to_csv('INT-lasAnimasCO(Cleaned).csv', index=False)
```
